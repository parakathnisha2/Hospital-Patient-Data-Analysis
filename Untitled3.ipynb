```python
# ==========================================
# HOSPITAL PATIENT DATA ANALYSIS PROJECT
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------
# STEP 1 : CREATE DATASET
# ------------------------------------------

data = {
    "Patient_ID": [101,102,103,104,105,106,107,108,109,110,111,112],
    "Name": ["Arun","Priya","Kumar","Divya","Ravi","Anu",
             "Vijay","Meena","Suresh","Nisha","Hari","Latha"],
    "Age": [25,40,35,60,15,50,30,70,45,20,55,28],
    "Disease": ["Fever","Diabetes","Fever","Heart Disease",
                "Fever","Diabetes","Asthma","Heart Disease",
                "Asthma","Fever","Diabetes","Asthma"],
    "Admission_Date": [
        "2025-01-10","2025-01-15","2025-02-05",
        "2025-02-12","2025-03-01","2025-03-10",
        "2025-04-08","2025-04-20","2025-05-05",
        "2025-05-18","2025-06-12","2025-06-25"
    ]
}

df = pd.DataFrame(data)

print("========== ORIGINAL DATA ==========")
print(df)

# ------------------------------------------
# STEP 2 : DATA CLEANING
# ------------------------------------------

df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

print("\n========== CLEANED DATA ==========")
print(df)

# ------------------------------------------
# STEP 3 : DISEASE WISE ANALYSIS
# ------------------------------------------

disease_count = df["Disease"].value_counts()

print("\n========== DISEASE WISE COUNT ==========")
print(disease_count)

# ------------------------------------------
# STEP 4 : AGE GROUP ANALYSIS
# ------------------------------------------

def age_group(age):
    if age <= 18:
        return "Child"
    elif age <= 40:
        return "Adult"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(age_group)

age_count = df["Age_Group"].value_counts()

print("\n========== AGE GROUP ANALYSIS ==========")
print(age_count)

# ------------------------------------------
# STEP 5 : PEAK ADMISSION ANALYSIS
# ------------------------------------------

df["Admission_Date"] = pd.to_datetime(df["Admission_Date"])

df["Month"] = df["Admission_Date"].dt.month_name()

monthly_admissions = df["Month"].value_counts()

print("\n========== MONTHLY ADMISSIONS ==========")
print(monthly_admissions)

peak_month = monthly_admissions.idxmax()

print("\nPeak Admission Month :", peak_month)

# ------------------------------------------
# STEP 6 : CHARTS IN SAME ROW
# ------------------------------------------

fig, axes = plt.subplots(1, 3, figsize=(18,5))

# BAR CHART
disease_count.plot(
    kind='bar',
    ax=axes[0],
    title='Disease Wise Count'
)
axes[0].set_xlabel("Disease")
axes[0].set_ylabel("Patients")

# PIE CHART
age_count.plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=axes[1]
)
axes[1].set_title("Age Group Distribution")
axes[1].set_ylabel("")

# LINE CHART
monthly_admissions.sort_index().plot(
    kind='line',
    marker='o',
    ax=axes[2],
    title='Admission Trend'
)
axes[2].set_xlabel("Month")
axes[2].set_ylabel("Admissions")
axes[2].grid(True)

plt.tight_layout()
plt.show()

```
